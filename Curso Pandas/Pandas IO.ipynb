# Selecionando Cursos

```python
nomes.sample(5)
```

```python
todas_matriculas = []
x = np.random.rand(20)
prob = x /sum(x)
```

```python
for index, row in nomes.iterrows():
    id = row.id_alunos
    matriculas = row.matriculas
    for i in range(matriculas):
        mat = [id, np.random.choice(cursos.index, p = prob)]
        todas_matriculas.append(mat)

matriculas = pd.DataFrame(todas_matriculas, columns = ['id_alunos', 'id_curso'])
```

```python
matriculas.head()
```

```python
matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns = {'id_alunos':'quantidade_alunos'})
```

```python
nomes.sample(3)
```

```python
cursos.head()
```

```python
matriculas.head()
```

```python
matriculas_por_curso = matriculas.groupby('id_curso').count().join(cursos['nome_do_curso']).rename(columns = {'id_alunos':'quantidade_alunos'})
```
